```python
from random import *
from time import *
import matplotlib.pyplot as plt
```

```python
def create_input(n):
    with open("input.txt","w") as data:
        li =[0]*n
        for i in range(n):
           li[i]=str(randint(1,100))+"\n"
        data.writelines(li)

    with open("input.txt","r") as data:
        l = data.readlines()
        s = list(map(str.strip, l))
        res = [eval(i) for i in s]
    return res
```

```python
def insertion_sort(lis):
    for i in range(1,len(lis)):
        key = lis[i]
        j=i-1
        while j>=0 and lis[j]>key:
            lis[j+1] = lis[j]
            j=j-1
        lis[j+1]=key
    endt =time()
    return lis
```

```python
def merge(arr, l, m, r):
    n1 = m - l + 1
    n2 = r - m

    L = [0] * (n1)
    R = [0] * (n2)

    for i in range(0, n1):
        L[i] = arr[l + i]

    for j in range(0, n2):
        R[j] = arr[m + 1 + j]

    i = 0
    j = 0
    k = l

    while i < n1 and j < n2:
        if L[i] <= R[j]:
            arr[k] = L[i]
            i += 1
        else:
            arr[k] = R[j]
            j += 1
        k += 1

    while i < n1:
        arr[k] = L[i]
        i += 1
        k += 1

    while j < n2:
        arr[k] = R[j]
        j += 1
        k += 1

def mergeSort(arr, l, r):
    if l < r:
        m = l+(r-l)//2
        mergeSort(arr, l, m)
        mergeSort(arr, m+1, r)
        merge(arr, l, m, r)
    return arr
```

```python
time1,time2 = [0]*1000,[0]*1000
x_points = list(range(1, 1001))
for i in range(1,1000):
    inp = create_input(i)
    bgt=time()
    insertion_sort(inp)
    endt=time()
    time1[i]=endt-bgt
    bgt=time()
    mergeSort(inp,0,i-1)
    endt=time()
    time2[i]=endt-bgt
plt.plot(x_points,time1,"r",label="Insertion-Sort")
plt.plot(x_points,time2,"b",label="Merge-Sort")
plt.legend()
plt.show()
```
